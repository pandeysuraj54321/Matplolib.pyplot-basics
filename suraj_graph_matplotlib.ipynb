```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,3,4,6,8,9]
y=[2,3,4,5,10,11]
plt.plot(x,y,label='2x')
plt.xlabel('AGE',fontdict={'fontsize':19,'color':'blue'},loc='right')
plt.ylabel('HEIGHT',fontdict={'fontsize':35,'color':'y'},loc='top')
plt.title('Variation of Height with Age ',fontdict={'fontsize':33,'color':'red','horizontalalignment':'right'})
plt.legend()
plt.xticks([2,3,5,6,8,9,13])
plt.yticks([2,3,4,5,6,7,8,9,10])
plt.show()
```

```python
x=[1,3,4,6,8,9]
y=[2,3,4,5,10,11]
plt.plot(x,y,label='8x',color='g',linewidth=5,marker='.',markersize=23,markeredgecolor='b')
plt.xlabel('AGE',fontdict={'fontsize':19,'color':'blue'},loc='right')
plt.ylabel('HEIGHT',fontdict={'fontsize':20,'color':'y'},loc='top')
plt.title('Variation of Height with Age ',fontdict={'fontsize':25,'color':'red','horizontalalignment':'center'})
plt.legend()
plt.xticks([2,3,5,6,8,9,13])
plt.yticks([2,3,4,5,6,7,8,9,10])
plt.show()
```

```python
x=[1,3,4,6,8,9]
y=[2,3,4,5,10,11]
plt.plot(x,y,label='8x',color='g',linewidth=4,marker='*',markersize=20,markeredgecolor='b',linestyle='--')
plt.xlabel('AGE',fontdict={'fontsize':19,'color':'blue'},loc='right')
plt.ylabel('HEIGHT',fontdict={'fontsize':20,'color':'y'},loc='top')
plt.title('Variation of Height with Age ',fontdict={'fontsize':25,'color':'red','horizontalalignment':'center'})
plt.legend()
plt.xticks([2,3,5,6,8,9,13])
plt.yticks([2,3,4,5,6,7,8,9,10])
plt.show()
```

```python
x=[1,3,4,6,8,9]
y=[2,3,4,5,10,11]
plt.plot(x,y,'g.-',label='8x')
x2=[1,2,3,4,5]
y2=[1,4,9,14,25]
plt.plot(x2,y2,'y*-',label='x**2')
plt.xlabel('AGE',fontdict={'fontsize':19,'color':'blue'},loc='right')
plt.ylabel('HEIGHT',fontdict={'fontsize':20,'color':'y'},loc='top')
plt.title('Variation of Height with Age ',fontdict={'fontsize':25,'color':'red','horizontalalignment':'center'})
plt.legend()


plt.show()
```

```python
plt.figure(figsize=(3,2),dpi=200)
x=[1,3,4,6,8,9]
y=[2,3,4,5,10,11]
plt.plot(x,y,'g.-',label='8x')
x2=[1,2,3,4,5]
y2=[1,4,9,14,25]
plt.plot(x2[:3],y2[:3],'y*-',label='x**2')
plt.plot(x2[2:],y2[2:],'y*--',label='x**2')
plt.xlabel('AGE',fontdict={'fontsize':19,'color':'blue'},loc='right')
plt.ylabel('HEIGHT',fontdict={'fontsize':20,'color':'y'},loc='top')
plt.title('Variation of Height with Age ',fontdict={'fontsize':25,'color':'red','horizontalalignment':'center'})
plt.legend()


plt.show()
plt.savefig('mygraph.png')
```

###  BAR CHART

```python
labels=['A','B','C']
values=[3,5,9]
plt.title('my first graph')
plt.xlabel('category')
plt.ylabel('count')
plt.bar(labels,values)

plt.show()
```

```python
labels=['A','B','C']
values=[3,5,9]
plt.title('my first graph')
plt.xlabel('category')
plt.ylabel('count')
bars=plt.bar(labels,values)
bars[0].set_hatch('o')
bars[2].set_hatch('--')
bars[1].set_hatch('*')
plt.show()
```

```python
labels=['A','B','C']
values=[3,5,9]
bars=plt.bar(labels,values)
patterns=['/','*','o']
for bar in bars:
    bar.set_hatch(patterns.pop())

plt.show()
```
